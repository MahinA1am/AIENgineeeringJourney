**Packakge Import**

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
nltk.download(['stopwords','punkt','wordnet'])
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize,word_tokenize

en_stopwords = stopwords.words('english')
```

**Stopwords removal**

```python
sentence = 'I am too old to work now and so i am not going to work today'
sentence_no_stopwords = ' '.join([word for word in sentence.split() if word not in en_stopwords])
sentence_no_stopwords
```

**REGEX**

```python
import re
sentence1 = r"kante bator is not a man"
print(re.search("kante",sentence1))
print(re.sub("kante","mante",sentence1))

reviews = [
    "Sara is very efficient and always meets deadlines.",
    "Namo is a great team player and helps others readily.",
    "Kita's attention to detail is impressive.",
    "Sara consistently goes the extra mile.",
    "Namo has a positive attitude that is contagious.",
    "Kita is a reliable and valuable asset to the team.",
    "Sara is a strong communicator and keeps everyone informed."
]

#All sara revies
sara_reviews = []
pattern_to_find = r'Sarah?'
for review in reviews:
  if re.search(pattern_to_find,review):
    sara_reviews.append(review)

for review in sara_reviews:
  #print(review, end = ' ')
  print(review)

#Review starting with K
k_reviews = []
pattern_to_find = r'^K'
for review in reviews:
  if re.search(pattern_to_find,review):
    k_reviews.append(review)

print(k_reviews)

#Review ending with e
e_reviews = []
pattern = r"e.$"
for r in reviews:
  if re.search(pattern,r):
    e_reviews.append(r)

print(e_reviews)

#Review with meets/helps/keeps
mkh_reviews = []
pattern = r"(meet|keep|help)s"
for r in reviews:
  if re.search(pattern,r):
    mkh_reviews.append(r)

print(mkh_reviews)

#Review without puncuation
npunc_reviews = []
pattern = r"[^\w\s]"
for r in reviews:
  new_r = re.sub(pattern,'',r)
  npunc_reviews.append(new_r)

print(npunc_reviews)
```

**Tokenize**

```python
lower_reviews = [r.lower() for r in reviews]
for r in lower_reviews:
  print(sent_tokenize(r))
for r in lower_reviews:
  print(word_tokenize(r))
```

**Stemming**

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer( )

tokens = ['likes','dislikes','connect','connected','connectivity','learn','learnt','learning','learners','learner','arise','complex','complexity','rise']

for t in tokens:
  print(t, ': ', ps.stem(t))
```

**Lemmatizing**

```python
from nltk.stem import WordNetLemmatizer
lm = WordNetLemmatizer()

for t in tokens:
  print(t, ': ', lm.lemmatize(t))
```

**Practice using real datas**et

```python
#Loading Dataset
df = pd.read_csv('tripadvisor_hotel_reviews.csv')

#Dropping Unwanted Column
df = df.drop(columns=['Rating'], axis = 1)

#PreProcessing
df['LowerReview'] = df['Review'].str.lower()
df['StopWordsReview'] = df['LowerReview'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
df['PuncReview'] = df['StopWordsReview'].apply(lambda x: re.sub(r'[*]','star',x))
df['PuncReview'] = df['PuncReview'].apply(lambda x: re.sub(r'[^\w\s]','',x))
df['TokenizeReview'] = df['PuncReview'].apply(lambda x: word_tokenize(x))
df['StemmedReview'] = df['TokenizeReview'].apply(lambda x: [ps.stem(word) for word in x])
df['LemmatizedReview'] = df['TokenizeReview'].apply(lambda x: [lm.lemmatize(word) for word in x])

#df.head()

#NGram
token_list = sum(df['LemmatizedReview'],[])

unigram = pd.Series(nltk.ngrams(token_list,1)).value_counts()
bigram = pd.Series(nltk.ngrams(token_list,2)).value_counts()

print(unigram.head())
print(bigram.head())

unigram[:10].sort_values().plot.barh(color = "green")
plt.title("Unigram Tokenized Occurence")
plt.show()

bigram[:10].sort_values().plot.barh(color = "purple")
plt.title("Bigram Tokenized Occurence")
plt.show()
```
